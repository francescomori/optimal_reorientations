```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import numpy.random as r
from tqdm import tqdm
from scipy.optimize import minimize_scalar
from functools import partial
import math
```

## Non-instantaneous reorientations

```python
def Vp(theta,thetaa,a):
    theta=modulo_2pi(theta)
    if np.abs(theta)>thetaa:
        return -a*(np.sign(theta))
    return 0
def modulo_2pi(num):
    result = num % (2 * math.pi)
    if result > math.pi:
        result -= 2 * math.pi
    elif result <= -math.pi:
        result += 2 * math.pi
    return result
```

```python
# Define the objective function you want to minimize
def F(x,Pe):
    x=modulo_2pi(x)
    return -np.sin(x)/(x+(1-np.exp(-Pe*(np.pi-x)))/Pe)

def v(Pe):
    #optimal speed
    
    F_fixed = partial(F, Pe=Pe)
    
    # Perform the minimization
    result = minimize_scalar(F_fixed, bounds=(0, np.pi+0.01), method='bounded')
    
    
    return max([0,-result.fun,-F(np.pi,Pe)])

def ps(Pe):
    #optimal speed
    
    thetaa=theta_a(Pe)
    
    # Perform the minimization
    
    
    
    return 1- Pe*thetaa/(1+Pe*thetaa-np.exp(Pe*(thetaa-np.pi)))

def theta_a(Pe):
    #optimal activation angle
    F_fixed = partial(F, Pe=Pe)
    
    # Define the interval over which you want to minimize the function
    
    # Perform the minimization
    result = minimize_scalar(F_fixed, bounds=(0, np.pi+0.01), method='bounded')
    
    if result.fun<0:
        return result.x
    else:
        return 4
    
def modulo_2pi(num):
    result = num % (2 * math.pi)
    if result > math.pi:
        result -= 2 * math.pi
    elif result <= -math.pi:
        result += 2 * math.pi
    return result
```

```python
t_f=10000#total time
dt=0.001#timestep


    
zs_sims=np.logspace(-2,2,6)
v_sims=[]
p_sims=[]

for z in tqdm(zs_sims):
    thetaa=theta_a(1/z)
    t=0
    x=0
    theta=0
    t_on=0
    #Langevin dynamics
    while t<t_f:
        theta=modulo_2pi(theta+np.sqrt(2*dt)*r.normal()+Vp(theta,thetaa,1/z)*dt)
        if np.abs(theta)<thetaa:
            x=x+np.cos(theta)*dt
        else:
            t_on+=dt
        t=t+dt

    v_sims.append(x/t_f)
    p_sims.append((t_on)/t_f)
```

```python
zs=np.logspace(-3,3,1000)
vs=[v(1/z) for z in zs]
plt.plot(zs,vs,linewidth=2.5,color='blue',label="Theory", zorder=1)
plt.scatter(zs_sims,v_sims,marker='x',color='red',s=70,label="Simulations", zorder=2)
plt.xscale("log")
plt.legend(fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel(r'$1/{\rm Pe}$',fontsize=30)
plt.ylabel(r'$v^*/v_0$',fontsize=30)
plt.tight_layout()
```

```python
zs=np.logspace(-3,3,1000)
thetaas=[theta_a(1/z)/np.pi for z in zs]
plt.plot(zs,thetaas,linewidth=2.5,color='blue')
plt.xscale("log")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel(r'$1/{\rm Pe}$',fontsize=30)
plt.ylabel(r'$\theta_a/\pi$',fontsize=30)
plt.tight_layout()
```

```python
zs=np.logspace(-3,3,1000)
vs=[ps(1/z) for z in zs]
plt.plot(zs,vs,linewidth=2.5,color='blue',label="Theory", zorder=1)
plt.scatter(zs_sims,p_sims,marker='x',color='red',s=70,label="Simulations", zorder=2)
plt.xscale("log")
plt.legend(fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel(r'$1/{\rm Pe}$',fontsize=30)
plt.ylabel(r'$p_r$',fontsize=30)
plt.tight_layout()
```
